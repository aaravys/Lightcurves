# Here, we import some key modules.

```python
import numpy as np
from astropy.time import Time
import pandas
import os 
import matplotlib.pyplot as plt
```

# Now, we open our csv file. This is important so we can look at the different headers for later on in the code.

```python
df = pandas.read_csv('data.csv')
```

```python
df.head()
```

# Now, we take a closer look at our x values, which are in mdj(Modified Julian Date.)

```python
print(df['mjd'])
```

# Now, we are plotting the graph. This graph represents the light curves from the star. 
#we use plt here. We set the figure size, the x and y axis values, and the plot title. 

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (20, 10))
plt.scatter(df['mjd'], df["flux"])
plt.title("S3TIC10000026150", size = "xx-large")
plt.xlabel("MJD(Modified Julian Date)", size="xx-large")
plt.ylabel("Flux", size="xx-large")
plt.grid(True)

plt.show()
```
